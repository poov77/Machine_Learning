```python
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv("train.csv")
train
```

```python
train.shape
```

```python
train.isna().sum()
```

```python
train.dropna(inplace = True)
```

```python
train.isna().sum()
```

#### Assumption 1: Test for Linearity

```python
sns.lmplot(x='x',y='x',data=train)
plt.title('x Vs y')
plt.show()
```

### 2. Test for Normality

```python
sns.distplot(a=train['x'],hist = True)
plt.show()
```

```python
from scipy import stats
stats.probplot(x=train['x'],dist='norm',plot=plt)
plt.show()
```

#### Model Buliding 

```python
x = train.drop(labels=['y'],axis = 1)
y = train[['y']]
```

```python
x
```

```python
y
```

#### Model Training 

```python
from sklearn.linear_model import LinearRegression
linear_model_1 = LinearRegression()
linear_model_1.fit(x,y)
```

```python
y_pred = linear_model_1.predict(x)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

#### Model Evaluation


```python
error = y - y_pred
error
```

# 5. Test for Homoscesdasticity

```python
plt.scatter(x = train['x'], y=error)
plt.title('Homoscedsticity Check - x Vs Error')
plt.xlabel('X')
plt.ylabel('Error')
plt.show()
```
