```python
x=d1[['HP','VOL','SP','WT']]
x
```

```python
y=d1[['MPG']]
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mlr=LinearRegression()
mlr.fit(x,y)
y_pred=mlr.predict(x)
```

```python
y_pred
```

```python
mlr.coef_
```

```python
mlr.intercept_
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,y_pred)
```

```python
error=y-y_pred
error
```

```python
plt.scatter(x['HP'],error)
plt.show()
```

```python
plt.scatter(x['VOL'],error)
plt.show()
```

```python
plt.scatter(x['SP'],error)
plt.show()
```
