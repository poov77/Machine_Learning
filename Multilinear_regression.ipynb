```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Example dataset (replace with your own x, y)
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([45000, 50000, 60000, 80000, 110000])

# Step 1: Transform features into polynomial terms
poly = PolynomialFeatures(degree=5)
x_poly = poly.fit_transform(x)

# Step 2: Fit the regression model
model = LinearRegression()
model.fit(x_poly, y)

# Step 3: Predict using the trained model
x_sorted = np.sort(x, axis=0)
y_pred_sorted = model.predict(poly.transform(x_sorted))

# Step 4: Plot
plt.scatter(x, y, color='red', label='Actual Salary Data')
plt.plot(x_sorted, y_pred_sorted, color='green', label='Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
