```python
from sklearn.model_selection import train_test_split
```

```python
# Select all rows, but only columns 0,1,3,4,5
x=df.iloc[:, [0, 2, 3, 4]]
x
```

```python
y=df.iloc[:,1:2]
y
```

```python
from sklearn.preprocessing import MinMaxScaler

ss = MinMaxScaler()
x = ss.fit_transform(x)
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=12)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=3)

# Fit on training data
knn.fit(x_train, y_train)

# Predictions
y_train_pred = knn.predict(x_train)
y_test_pred  = knn.predict(x_test)
y_train_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_train, y_train_pred)
r2 = r2_score(y_train, y_train_pred)

print("MSE:", mse)
print("R²:", r2)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Create and fit the regressor
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(x_train, y_train)

# Predict continuous values
y_train_pred = knn.predict(x_train)

# Evaluate with regression metrics
mse = mean_squared_error(y_train, y_train_pred)
r2 = r2_score(y_train, y_train_pred)

print("MSE:", mse)
print("R²:", r2)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
print("RMSE:", rmse)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_train, y_train_pred)
print("MAE:", mae)
```

```python
import numpy as np

tolerance = 10  # acceptable error margin
accuracy = np.mean(np.abs(y_train - y_train_pred) <= tolerance)
print("Custom Accuracy:", accuracy)
```
