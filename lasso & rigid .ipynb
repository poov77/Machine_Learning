```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge, Lasso
from sklearn.preprocessing import StandardScaler

# 1. Create a dummy dataset matching our software engineer scenario
np.random.seed(42)
n_samples = 100

data = {
    'Years_Exp': np.random.randint(1, 15, n_samples),
    'Coffee_Cups': np.random.randint(0, 8, n_samples), # Noise
    'GitHub_Projects': np.random.randint(1, 30, n_samples)
}
df = pd.DataFrame(data)

# Create a redundant column (Years_Exp in months)
df['Exp_Months'] = df['Years_Exp'] * 12 

# Generate target salary with some noise
df['Salary'] = 50000 + (df['Years_Exp'] * 8000) + (df['GitHub_Projects'] * 500) + np.random.normal(0, 2000, n_samples)

# Separate features (X) and target (y)
X = df.drop(columns=['Salary'])
y = df['Salary']

# CRUCIAL: Regularization requires scaling first!
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Train Ridge (L2) and Lasso (L1)
# 'alpha' is the tuning parameter that controls the strength of the penalty
ridge = Ridge(alpha=10.0)
lasso = Lasso(alpha=100.0)

ridge.fit(X_scaled, y)
lasso.fit(X_scaled, y)

# 3. Print out the calculated weights (Coefficients)
feature_names = X.columns
print("--- RIDGE COEFFICIENTS (Shrinkage) ---")
for name, coef in zip(feature_names, ridge.coef_):
    print(f"{name}: {coef:.2f}")

print("\n--- LASSO COEFFICIENTS (Feature Selection) ---")
for name, coef in zip(feature_names, lasso.coef_):
    print(f"{name}: {coef:.2f}")
    
```
